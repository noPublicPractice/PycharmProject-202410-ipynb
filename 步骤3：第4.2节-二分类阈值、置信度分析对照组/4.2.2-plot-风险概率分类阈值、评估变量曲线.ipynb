```python
import json
import matplotlib.pyplot as plt
plt.style.use('_mpl-gallery')
plt.rcParams['font.family'] = ['SimHei', 'Times New Roman']
```

```python
class const:
    label_ch_list = ['房颤', '动脉粥样硬化', '心梗', '脑梗']
    label_en_list = ['AF', 'AS', 'MI', 'CI']
    density = 200
    threshold_with_assessed_values_path = r'..\..\PycharmProject-202412-集成学习\绘图数据【保留】\%s%s-threshold-with-acc-ppv-tpr-dict-density=%d.json'
```

```python
def plt_assessed_values(label_en, label_ch):
    now_json_path_file_name = const.threshold_with_assessed_values_path % (label_en, label_ch, const.density)
    assessed_values_dict = json.load(open(now_json_path_file_name, 'r', encoding='utf-8'))
    fig_3 = plt.figure(figsize=(16, 8))
    ax1 = fig_3.add_subplot(111)
    plt.plot(assessed_values_dict['classify_threshold'], assessed_values_dict['acc'], label='ACC', c='k', linewidth=4)
    plt.plot(assessed_values_dict['classify_threshold'], assessed_values_dict['ppv'], label='PPV', c='r', linewidth=4)
    plt.plot(assessed_values_dict['classify_threshold'], assessed_values_dict['tpr'], label='TPR', c='g', linewidth=4)
    plt.plot(assessed_values_dict['classify_threshold'], assessed_values_dict['f1'], label='F1', c='b', linewidth=4)
    plt.plot(assessed_values_dict['classify_threshold'], assessed_values_dict['acc_multiply_tpr'], label='ACC×TPR', c='y', linewidth=4)
    plt.rcParams.update({'font.size': 32})
    plt.legend(shadow=True, fancybox=True, bbox_to_anchor=(1.05, 0.3), loc=3, borderaxespad=0)
    plt.xlabel('风险概率分类阈值')
    plt.ylabel('评估值')
    ax1.set_title(label_ch)
    ax1.set_xlim([-0.02, 1.02])
    ax1.set_ylim([-0.02, 1.02])
    ax1.spines['bottom'].set_linewidth(3)  # 设置底部坐标轴的粗细
    ax1.spines['left'].set_linewidth(3)  # 设置左边坐标轴的粗细
    ax1.spines['right'].set_linewidth(3)  # 设置右边坐标轴的粗细
    ax1.spines['top'].set_linewidth(3)  # 设置上部坐标轴的粗细
    plt.show()
```

```python
def do_work():
    for label_en, label_ch in zip(const.label_en_list, const.label_ch_list):
        plt_assessed_values(label_en, label_ch)
if __name__ == '__main__':
    do_work()
```
